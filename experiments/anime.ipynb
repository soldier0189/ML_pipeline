```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
anime = pd.read_csv("anime.csv")
sys = pd.read_csv("anime_with_synopsis.csv")
```

```python
anime.head()
```

```python
sys.head()
```

```python
anime = anime.merge(sys,on="MAL_ID")
anime.head()
```

# .1 Columns

```python
# mal_id, name, genres, sypno
anime = anime[["MAL_ID","Name_x","Genres_x","sypnopsis","Score_y"]]
anime.head()
```

```python
anime["Tag"] = anime.Genres_x +" "+ anime.sypnopsis
```

```python
anime.head()
```

```python
anime.drop(columns=["Genres_x","sypnopsis"],inplace=True)
anime.head()
```

```python
anime.dropna(inplace=True)
```

```python
anime.Score_y = anime.Score_y.replace({
    "Unknown":np.nan
})
anime.Score_y.fillna(anime.Score_y.median,inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
from nltk.corpus import stopwords

stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))
def preprocess(text):
    # Lowercase
    text = text.lower()

    text = ''.join(char for char in text if char.isalnum() or char.isspace())
    
    words = text.split()

    words = [stemmer.stem(word) for word in words if word not in stop_words]
    
    return ' '.join(words)
anime.Tag = anime.Tag.apply(preprocess)    
tfidf = TfidfVectorizer(stop_words='english', max_features=10000)
tfidf_matrix = tfidf.fit_transform(anime['Tag'])

similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend(anime_name, top_n=5):
    # Find the index of the anime
   
    idx = anime[anime['Name_x'] == anime_name].index[0]
    
    # Get similarity scores for this anime with all others
    distances = similarity[idx]
    
    # Sort by similarity (highest first) and get top_n
    anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:top_n+1]
    
    # Print recommended anime names
    for i in anime_list:
        print(anime.iloc[i[0]]['Name_x'])
```

```python
recommend("One Punch Man")
```

```python
import pickle

# Save TF-IDF vectorizer
pickle.dump(tfidf, open("tfidf_vectorizer.pkl", "wb"))

# Save similarity matrix
pickle.dump(similarity, open("similarity_matrix.pkl", "wb"))

# Save cleaned DataFrame
anime.to_pickle("anime_df.pkl")
```

```python
import streamlit as st

tfidf = pickle.load(open("tfidf_vectorizer.pkl", "rb"))
similarity = pickle.load(open("similarity_matrix.pkl", "rb"))

# App title
st.title("Anime Recommender System")

# User input
anime_name_input = st.text_input("Enter your favorite anime:")

if st.button("Recommend"):
    anime_name = anime_name_input.replace(" ", "").lower()
    if anime_name in anime['Name_x'].values:
        idx = anime[anime['Name_x'] == anime_name].index[0]
        distances = similarity[idx]
        anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        st.write("Top recommendations:")
        for i in anime_list:
            st.write(anime.iloc[i[0]]['name'])
    else:
        st.write("Anime not found. Try another name.")

```
